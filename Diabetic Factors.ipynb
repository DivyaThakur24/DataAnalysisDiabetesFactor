# 2. Data Visualisation

```python
#Horizontal Bar Plot
dataset.plot.bar(stacked=True, fontsize=12)
```

```python
#Histogram
dataset.hist(figsize = (10,8))
```

```python
#Box Plot
dataset.plot(kind= 'box' , subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(10,8))
```

```python
#Correlation Plot
Corr = dataset[dataset.columns].corr()
sns.heatmap(Corr, annot=True)
```

# By Observing Outcome Ratios of Correlation:

1. Features Predominant - BMI, Age, Pregnancies (External Factors) Glucose, Insulin(Internal Factors) -- Descending Order
2. Features Not Critical - Skin Thickness, Blood Pressure

# Conclusions Drawn:

1. Keep your BMI in check to avoid High levels of Glucose and Insulin.
2. Keep your Glucose and Insulin levels in check as age progresses.
3. Pregnant Ladies, also shoul keep a close watch on their Glucose and Insulin Levels
